```python
def get_clustered_dataframe(incoming_df, column_list, sample_size):
    gmm = mixture.BayesianGaussianMixture(n_components=2, 
                                          covariance_type="full", 
                                          n_init=100, 
                                          random_state=42).fit(incoming_df)
    clustered_data = gmm.sample(sample_size)
    clustered_df = pd.DataFrame(clustered_data[0], columns=column_list)
    return clustered_df
```

```python
data = data.drop(['mu0', 'x1', 'x16', 'x20'], axis=1)
```

```python
output_dict2 = notears.notears_standard(
                          data, 
                          notears.loss.least_squares_loss, 
                          notears.loss.least_squares_loss_grad, 
                          e=1e-8, 
                          verbose=False)
```

```python
print('Acyclicity loss: {}'.format(output_dict2['h']))
print('Least squares loss: {}'.format(output_dict2['loss']))
```

```python
clustered_df = get_clustered_dataframe(data, list(data), 750)
```

```python
output_dict3 = notears.notears_standard(
                          clustered_df, 
                          notears.loss.least_squares_loss, 
                          notears.loss.least_squares_loss_grad, 
                          e=1e-8, 
                          verbose=False)
print('Acyclicity loss: {}'.format(output_dict3['h']))
print('Least squares loss: {}'.format(output_dict3['loss']))
```
