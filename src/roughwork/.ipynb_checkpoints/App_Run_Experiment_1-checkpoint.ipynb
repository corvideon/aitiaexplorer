Now look at the results metrics...

```python
target_graph_str = TargetData.simulated_data_1_graph()
causal_graph_target = GraphUtil.get_causal_graph_from_dot(target_graph_str)
causal_graph_target.draw()
```

```python
target_graph = GraphUtil.get_nxgraph_from_dot(target_graph_str)
df_results = pd.DataFrame(columns=('Algo', 'Ind Rels', 'AUC', 'SHD', 'ISO'))
metrics = GraphMetrics()
target_ind_rels = len(causal_graph_target.get_all_independence_relationships())
for result in analysis:
    if result['dot_str'] is not None and result['causal_graph'] is not None:
        pred_graph = GraphUtil.get_nxgraph_from_dot(result['dot_str'])
        prec_recall = metrics.precision_recall(target_graph, pred_graph)
        shd = metrics.SHD(target_graph, pred_graph)
        isomorphic = nx.is_isomorphic(target_graph, pred_graph)
        new_row = {'Algo':result['algo_name'], 
                   'Target Ind Rels': 
                   'Ind Rels':result['num_ind_rel'], 
                   'AUC':prec_recall[0], 
                   'SHD':shd,
                   'ISO': isomorphic}
        df_results = df_results.append(new_row, ignore_index=True)
df_results
```
