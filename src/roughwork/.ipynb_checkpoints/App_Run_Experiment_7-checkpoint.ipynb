```python
import os
import sys
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from pycausal.pycausal import pycausal

module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)
from aitia_explorer.app import App

# stop the warning clutter
import warnings
warnings.filterwarnings('ignore')
```

```python
pc = pycausal()
pc.start_vm()
```

```python
aitia = App()
```

```python
scm2 = aitia.data.scm2()
scm2.cgm.draw()
```

```python
target_graph_str = str(scm2.cgm.draw())
```

```python
df = scm2.sample(1000)
df
```

```python
analysis_results, summary = aitia.run_analysis(df, 
                                               target_graph_str=target_graph_str, 
                                               algorithm_list=[
                                                   aitia.algo_runner.PC,
                                                   aitia.algo_runner.FCI
                                               ],
                                               pc=pc)
summary
```
